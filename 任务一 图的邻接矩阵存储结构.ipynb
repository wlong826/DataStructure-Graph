### 1. 定义图类

```python
class Graph():
    # 初始化函数
    def __init__(self, vexs):
        self.vexs = vexs # 存储图中顶点的数据
        n = len(vexs)
        self.matrix = [[0]*n for i in range(n)] # 二维数组，记录顶点之间的关系
    
    # 创建无向图, 返回邻接矩阵
    def CreateDN(self, relation):
        for rela in relation: # 遍历点与点之间关系
            m = rela[0]
            n = rela[1]
            self.matrix[m][n] = 1
            self.matrix[n][m] = 1 # 无向图的二阶矩阵沿主对角线对称
        return self.matrix
    
    # 创建有向图
    def CreatDG(self, relation):
        for rela in relation: # 遍历点与点之间关系
            m = rela[0]
            n = rela[1]
            self.matrix[m][n] = 1
        return self.matrix
    
    # 创建无向网。和无向图唯一的区别就是二阶矩阵中存储的是权值
    def CreateUDN(self, relation):
        for rela in relation:
            m = rela[0]
            n = rela[1]
            self.matrix[m][n] = rela[2]
            self.matrix[n][m] = rela[2]
        return self.matrix
    
    # 创建有向网，和有向图不同的是二阶矩阵中存储的是权值。
    def CreateUDG(self, relation):
        for rela in relation:
            m = rela[0]
            n = rela[1]
            self.matrix[m][n] = rela[2]
        return self.matrix
```

### 2. 创建无向图

![Image Name](https://cdn.kesci.com/upload/image/qwy3cr59cb.png?imageView2/0/w/400/h/400)

```python
vexnum = 4 # 顶点个数
V0, V1, V2, V3 = range(vexnum)
vexs = ['V0', 'V1', 'V2', 'V3'] # 顶点数据
relation_DN = [[V0,V1], [V0,V2], [V0, V3], [V1, V2], [V2, V3]] # 存储顶点之间关系

graph_DN = Graph(vexs)
# 无向图，返回邻接矩阵
matrix_DN = graph_DN.CreateDN(relation_DN)
for DN in matrix_DN:
    print(DN)
```

### 3. 创建有向图

![Image Name](https://cdn.kesci.com/upload/image/qwy3hzc8n2.png?imageView2/0/w/400/h/400)

```python
relation_DG = [[V0, V1], [V0, V3], [V1, V2], [V2, V0], [V3, V2]] # 存储顶点之间关系
graph_DG = Graph(vexs)
# 有向图，返回邻接矩阵
matrix_DG = graph_DG.CreatDG(relation_DG)
for DG in matrix_DG:
    print(DG)
```

### 4. 创建无向网

- 和无向图唯一的区别就是二阶矩阵中存储的是权值

![Image Name](https://cdn.kesci.com/upload/image/qwy3x8df4c.png?imageView2/0/w/400/h/400)

```python
relation_UDN = [[V0, V1, 2], [V0, V2, 5], [V0, V3, 10], [V1, V2, 4], [V2, V3, 6]] # 存储顶点之间关系
graph_UDN = Graph(vexs)
# 无向网，返回邻接矩阵
matrix_UDN = graph_UDN.CreateUDN(relation_UDN)
for UDN in matrix_UDN:
    print(UDN)
```

### 5. 创建有向网

- 和有向图不同的是二阶矩阵中存储的是权值；

![Image Name](https://cdn.kesci.com/upload/image/qwy4ij3pfd.png?imageView2/0/w/400/h/400)

```python
relation_UDG = [[V0, V1, 4], [V0, V3, 3], [V1, V2, 8], [V2, V0, 5], [V3, V2, 2]] # 存储顶点之间关系
graph_UDG = Graph(vexs)
# 有向网，返回邻接矩阵
matrix_UDG = graph_UDG.CreateUDG(relation_UDG)
for UDG in matrix_UDG:
    print(UDG)
```
